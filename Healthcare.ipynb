```python
X = data.drop('Outcome',axis = 1)
Y = data['Outcome']
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc=MinMaxScaler()
sc.fit(X)
new_data = sc.transform(X)
new_data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.25,random_state = 13)
```

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression()
log_reg.fit(x_train,y_train)
y_pred = log_reg.predict(x_test)
y_pred
```

```python
from sklearn import metrics
```

```python
acc_log = metrics.accuracy_score(y_pred,y_test)
```

```python
acc_log
```

```python
confusion_mat = metrics.confusion_matrix(y_pred,y_test)
confusion_mat
```

```python
classi_report = metrics.classification_report(y_pred,y_test)
```

```python
print(classi_report)
```

```python
log_reg_proba = log_reg.predict_proba(x_test)
proba1 =log_reg_proba[:,1]
fpr,tpr,thresh = metrics.roc_curve(y_test,proba1)
```

```python
auc_log = metrics.auc(fpr,tpr)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(dpi=90)
plt.title("AUC-ROC CURVE")
plt.xlabel("False positive rate")
plt.ylabel("True postive rate")
plt.plot(fpr,tpr,'red',label = "AUC score = %0.2f"%auc_log) 
plt.plot(fpr,fpr,'g--')
plt.legend()
```
